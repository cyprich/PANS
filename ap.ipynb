## Linearna regresia

```python
import numpy as np 
import matplotlib.pyplot as plt

# data 
f = np.array([1, 1, 3, 6, 9, 10])
t = np.linspace(0, len(f) - 1, len(f))

# funkcia: y(t) = a + bt
(a, b) = np.polyfit(t, f, 1)

plt.figure()
plt.title(f"""$y = a + bt$
$a = {a:.2f}$
$b = {b:.2f}$""")
plt.plot(t, f, label="data")

t = np.linspace(0, len(f) - 1, 1000)
f = [(i * a + i * b) for i in t]
plt.plot(t, f, label="aproximacia")
plt.legend()

pass
```

## Linearna regresia - maticovy tvar

Bez pouzitia funkcie `np.polyfit`, lebo to sa mi zdalo ako cheating

```python
import numpy as np 
import matplotlib.pyplot as plt

# data 
f = np.array([-10, 1, 3, 6, 90])
t = np.linspace(0, len(f) - 1, len(f))

# bazicka matica
B = np.array([
    np.pow(t, 0),
    np.pow(t, 1),
])

# vzorec: 
# fp = c * B
# c = (f * BT) * (B * BT)^-1

c = (
    np.dot(f, B.transpose()) 
    @ 
    np.linalg.inv(B @ B.transpose())
)

fp = c @ B

plt.figure()
plt.title(f"""$y = a + bt$
$a = {c[1]:.2f}$
$b = {c[0]:.2f}$""")
plt.plot(t, f, label="data")
plt.plot(t, fp, label="priemet")

plt.grid()
plt.legend()

pass
```
