Sigmoid

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def sig(z): 
    return 1 / (1 + np.exp(-z))


def grad_sig(z): 
    a = sig(z)
    return a * (1 - a)

steps = 1000
x = np.arange(-5, 5, 1/steps)

sig_y = np.array([sig(i) for i in x])
grad_sig_y = np.array([grad_sig(i) for i in x])

plt.figure()
plt.plot(sig_y, label="Sigmoid")
plt.plot(grad_sig_y, label="Gradient Sigmoid")
plt.legend()
plt.show()
```
