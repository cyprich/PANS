## 1. Výsledok jedného študenta

Začneme analýzou výsledku jedného študenta.
Máme:
- meno študenta
- počet získaných bodov
- maximálny počet bodov

```python
# TODO vytvoriť premenné pre info o študentovi
meno = "Jozko"
body = 70
max_body = 100
```

```python
# TODO vypočítať percentuálny podiel ktorý daný študent získal a vypísať ho do konzoly na 2 desatinné miesta
# Hint Na formátovaný výpis môžeme použiť fstring

percenta = body/max_body * 100
print(f"Student {meno} ziskal {percenta:.2f}%")
```

## 2. Prešiel / neprešiel

Študent prejde testom, ak dosiahne aspoň 61 %.
Na základe výsledku rozhodneme, či študent prešiel.

```python
# TODO
limit = 61

if percenta >= limit: 
    print("Presiel")
else: 
    print("Nepresiel")
```

## 5. Automatické hodnotenie študentov

Pre každého študenta vypíšeme:
- meno
- percentá
- informáciu, či prešiel

```python
# TODO
for (meno, body) in zip(mena, body): 
    print(f"{meno}, {body} bodov, {'presiel' if body/max_body * 100 >= limit else 'nepresiel'}")
```

## 6. Lepšia reprezentácia dát

Zoznamy `mena` a `body` patria k sebe.
Skúsime ich uložiť do jednej štruktúry.

```python
studenti = [
    {"meno": "Anna", "body": 78},
    {"meno": "Peter", "body": 55},
    {"meno": "Jana", "body": 92},
    {"meno": "Martin", "body": 61}
]
```

```python
# TODO vypis pre kazdeho studenta percenta a informaciu ci presiel alebo nie
for s in studenti: 
    print(f"{s["meno"]} {'presiel' if s["body"] >= 61 else 'nepresiel'}")
```

## 7. Funkcia na vyhodnotenie študenta

Aby sme nemuseli rovnaký kód písať stále dokola,
vytvoríme funkciu.

```python
# TODO metoda na vyhodnotenie uspesnosti studenta
def presiel(body: int) -> str: 
    return "Presiel" if body >= limit else "Nepresiel"
```

```python
# TODO modifikacia vypisu s vlastnou metodou
for s in studenti: 
    print(f"{s["meno"]} {presiel(s["body"])}")
```

## 8. Objektové programovanie v pythone

```python
class Student:
    def __init__(self, meno, body, maximum=100):
        self.meno = meno
        self.body = body
        self.maximum = maximum

    def percenta(self):
        return self.body / self.maximum * 100

    def vyhodnot(self, hranica=60):
        if self.percenta() >= hranica:
            return "prešiel"
        else:
            return "neprešiel"
```

```python
studenti = [
    Student("Anna", 78),
    Student("Peter", 55),
    Student("Jana", 92),
    Student("Martin", 61)
]
```

```python
for s in studenti:
    print(f"{s.meno}: {s.percenta():.1f}% – {s.vyhodnot()}")
```
